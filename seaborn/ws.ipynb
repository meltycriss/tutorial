```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# s = pd.Series(['orca', 0 , 2.1, 0., .5], index=entry)
# data = data.append(s, ignore_index=True)
# self.data.to_csv('{}/train_data{}.csv'.format(dir, suffix))
# df = pd.read_csv('{}/{}/{}'.format(dir, subfolder, csv_name)).drop(['Unnamed: 0'], axis=1)
# plt.figure()
# plt.savefig('{}/{}'.format(dir, name), dpi=200)
# plt.close()

# initialize dataframes
epi_entry = ['Setting', 'Test index', 'Efficiency', 'Connectivity', 'Smoothness']
epi_title = {e:[] for e in epi_entry}
epi_data = pd.DataFrame(epi_title)

sum_entry = ['Setting', 'Success rate']
sum_title = {e:[] for e in sum_entry}
sum_data = pd.DataFrame(sum_title)

# collecting data
setting = ['Adapted ORCA', 'Independent PPO', 'Our method', 'Dummy']
num_test = 10
for s in setting:
    list_efficiency = []
    list_connectivity = []
    list_smoothness = []
    success_rate = 0
    for n in range(num_test):
        list_efficiency.append(np.random.uniform(1, 2))
        list_connectivity.append(np.random.uniform(0, 0.5))
        list_smoothness.append(np.random.uniform(0, 1))
        success_rate += 1 if np.random.rand() < .5 else 0
    # episodic metrics
    list_setting = [s for n in range(num_test)]
    list_test_index = list(range(num_test))
    
    frame = [list_setting, list_test_index, list_efficiency, list_connectivity, list_smoothness]
    frame = list(map(list, zip(*frame)))
    
    df = pd.DataFrame(frame, columns=epi_entry)
    epi_data = epi_data.append(df, ignore_index=True)
    
    # summarized metrics
    success_rate /= 1. * num_test
    df = pd.Series([s, success_rate], index=sum_entry)
    sum_data = sum_data.append(df, ignore_index=True)
```

```python
# color stuff
colors = ["salmon pink", "tiffany blue", "light lavender", "peach", "light blue grey", "pastel yellow", "foam green",  ]
sns.set_palette(sns.xkcd_palette(colors))
sns.palplot(sns.xkcd_palette(colors))
```

```python
sns.barplot(x='Setting', y='Connectivity', data=epi_data, capsize=.1)
```

```python
sns.barplot(x='Setting', y='Success rate', data=sum_data)
```

```python
# convergence
convergence_entry = ['Test index', 'Iteration', 'Accumulated total rewards']
convergence_title = {e:[] for e in convergence_entry}
convergence_data = pd.DataFrame(convergence_title)
num_convergence = 5
num_iteration = 1000
for n in range(num_convergence):
    list_iteration = []
    list_reward = []
    for it in range(num_iteration):
        if it % 10 == 0:
            list_iteration.append(it)
            list_reward.append(np.random.uniform(6, 10))
    list_test_index = [n for _ in range(len(list_iteration))]

    frame = [list_test_index, list_iteration, list_reward]
    frame = list(map(list, zip(*frame)))
    
    df = pd.DataFrame(frame, columns=convergence_entry)
    convergence_data = convergence_data.append(df, ignore_index=True)

sns.lineplot(x="Iteration", y="Accumulated total rewards", data=convergence_data)
```
